```python
from tkinter import *
from tkinter import ttk
from tkcalendar import DateEntry
from tkinter import messagebox
from tkinter import Menu
import pandas as pd
import openpyxl
import pprint


sc = Tk()
sc.title('NT POS')
method = IntVar()
method.set(1)
value = ''
total_price=''
order_no = 0 

s = ttk.Style()
s.configure('new.TFrame', background='#FEFBE7')

tab_control = ttk.Notebook(sc)




tab1 = ttk.Frame(tab_control,style='new.TFrame')

def export_bill():
    
    bill = {"Item_name": [],
            "Price": [],
            "Quantity": [],
            "Voucher":[combo.get(),"","","",'end =""'],
            "Total": [total_print.cget('text'),"","","",'end=""'],
            "Date":[cal.get(),"","","",'end=""']}
    
    dem = 0
    if quantity_oolong.get() > 0:
        bill['Item_name'].append('Oolong Tea')
        bill['Price'].append('35')
        bill['Quantity'].append(quantity_oolong.get())
        dem += 1
    if quantity_fruit.get() > 0:
        bill['Item_name'].append('Fruit Tea')
        bill['Price'].append('45') 
        bill['Quantity'].append(quantity_fruit.get())
        dem += 1
    if quantity_black.get() > 0:
        bill['Item_name'].append('Black Coffee')
        bill['Price'].append('20') 
        bill['Quantity'].append(quantity_black.get())
        dem += 1
    if quantity_condensed.get() > 0:
        bill['Item_name'].append('Coffee with condensed milk')
        bill['Price'].append('25') 
        bill['Quantity'].append(quantity_condensed.get())   
        dem += 1
        
    bill['Voucher'] = [combo.get()] + [""]*(dem -1)
    bill['Total'] = [total_print.cget('text')] + [""]*(dem -1)
    bill['Date'] = [cal.get()] + [""]*(dem -1)
    export = pd.DataFrame(bill)
    export.to_excel(r'D:\Codegym\Bill.xlsx', index = False) 
    
    f = open('Bill_2.txt','a')
    f.write('\n-------------------------------------NT Coffee-------------------------------------\n')  
    f.write(str(export))  
    
    f.close()
    print(export)
noidung_et = StringVar()
date_time = StringVar()
combo = ttk.Combobox(tab1, values=['NONE','HELLOAUG', 'NTCF', 'NICEAUTUM'], width = 31,font =("Arial",20),justify ='center')

combo.grid(row = 7, column = 2,sticky = 'W',padx =(10,0))



quantity_oolong = IntVar()
quantity_fruit = IntVar()
quantity_black = IntVar()
quantity_condensed = IntVar()
IntVar() == 0

def click_oolong():
    quantity_oolong.set(quantity_oolong.get() + 1)
def click_fruit():
    quantity_fruit.set(quantity_fruit.get() + 1)
def click_black():
    quantity_black.set(quantity_black.get() + 1)
def click_condensed():
    quantity_condensed.set(quantity_condensed.get() + 1) 
    
def clear():
    quantity_oolong.set(0)
    quantity_fruit.set(0)
    quantity_black.set(0)
    quantity_condensed.set(0)
    total_print.config(text = '0')
    combo.set("")
    payment_confirm.config(text = '')
    
    
def total_bill():
    total_price = (int(soluong_oolong.get())*35 + int(soluong_fruit.get())*45 + int(soluong_black.get())*20 + int(soluong_condensed.get())*25)*1000
    if combo.get() == 'HELLOAUG':
        total_price = int(total_price)*0.9
        
    elif combo.get() == 'NTCF' and total_price >= 100000:
        total_price = int(total_price)*0.8
        
    elif combo.get() == 'NICEAUTUM' and total_price >= 50000 :
        
        click_condensed()
        total_price = total_price
        
    elif combo.get() == '' or combo.get() == 'NONE':
        total_price = int(total_price)
        
    total_print.config(text = total_price)  
        
def payment():
        
        if total_print['text'] == '0' or total_price == '0':
            res = messagebox.showerror("Error", "Please choose atleast one item")
        else:
            res = messagebox.askyesno('Confirmation', 'Do you want to confirm the order?')
            if res == True: 
                
                res = messagebox.showinfo('Bill', 'Order Payment Successful!')
                
                export_bill()
                clear()
                
            elif res == False:
                payment_confirm.config(text = '')
                    
def exit():
    res = messagebox.askyesno('Confirmation', 'Do you want to exit NT POS?')
    if res == True:
        sc.destroy()
def click_method():
    if value == 1:
        payment_cash.config(state = 'normal')
        payment_card.config(state = 'disabled')
        payment_e_wallet.config(state = 'disabled')
        
    elif value == 2:
        payment_cash.config(state = 'disabled')
        payment_card.config(state = 'normal')
        payment_e_wallet.config(state = 'disabled')
       
    elif value == 3:
        payment_cash.config(state = 'disabled')
        payment_card.config(state = 'disabled')
        payment_e_wallet.config(state = 'normal')
    
   
    
cal=DateEntry(tab1, selectmode='datetime',textvariable=noidung_et,font =("Arial",14),justify ='center')
cal.grid(column = 3,row = 14,sticky ='E',padx =(0,30))


ten_mon = Label(tab1, text = 'Name',font = ("Arial",14), height=1, width=30,bg='#FEFBE7',justify ='center')
gia = Label(tab1, text = 'Price (x1000)(VND)',font = ("Arial",14), height=1, width=30,justify ='center',bg='#FEFBE7')
soluong = Label(tab1, text = 'Quantity',font = ("Arial",14), height=1, width=30,justify ='center',bg='#FEFBE7')
ma_nhanvien = Label(tab1, text = 'Staff ID: NamTran',font = ("Arial",14), height=1, width=20,bg='#FEFBE7', justify ='right')
voucher = Label(tab1, text = 'Voucher',font =("Arial",20,'bold'),bg='#FEFBE7',fg ='#F73D93')


ten_mon.grid(row = 1, column = 1,sticky ='W',pady = (0,10))
gia.grid(row = 1, column = 2,pady = (0,10))
soluong.grid(row = 1, column = 3,pady = (0,10))
ma_nhanvien.grid(row =13, column = 3,sticky ='E')
voucher.grid(row =7, column = 1)

Button(tab1,text = 'Oolong Tea',font = ("Arial",14),fg='#A10035',bg='#FEFBE7', height=1,justify ='center', width=30,command = click_oolong).grid(row = 2, column = 1,sticky ='W')
Button(tab1,text = 'Fruit Tea',height=1,font = ("Arial",14),fg='#A10035',bg='#FEFBE7', width=30,command =click_fruit).grid(row = 3, column =1,sticky ='W')
Button(tab1,text = 'Black Coffee',height=1,font = ("Arial",14),fg='#A10035',bg='#FEFBE7', width=30,command = click_black).grid(row = 4, column =1,sticky ='W')
Button(tab1,text = 'Coffee with condensed milk',height=1,fg='#A10035',bg='#FEFBE7',font = ("Arial",14), width=30,command = click_condensed).grid(row = 5, column =1,sticky ='W')
Button(tab1,text = 'Payment Confirmation',font = ("Arial",14,'bold'), height=2,fg ='#00EAD3',bg='#FEFBE7', width=20, command = payment).grid(row = 12, column = 2,pady = (20,0))
Button(tab1,text = 'Clear',font = ("Arial",14,'bold'), height=1, width=10,bg='#FEFBE7', command = clear, fg = '#3DB2FF').grid(row = 6, column = 3, pady =(10,10))
Button(tab1,text = 'EXIT',font = ("Arial",14,'bold'), height=1,bg='#FEFBE7', width=10,command = exit,fg='red').grid(row = 14, column = 1,sticky ='W')
total = Button(tab1,text = "Total:", width = 5, height = 1,bg='#FEFBE7',font =("Arial",20), command = total_bill)
total.grid(row = 8, column = 1,pady =(10,0))

gia_oolong = Label(tab1, text = '35',font = ("Arial",14), height=1, width=20,fg='#A10035',background='#FEFBE7')
gia_fruit = Label(tab1, text = '45',font = ("Arial",14), height=1, width=20,fg='#A10035',background='#FEFBE7')
gia_black = Label(tab1, text = '20',font = ("Arial",14), height=1, width=20,fg='#A10035',background='#FEFBE7')
gia_condensed = Label(tab1, text = '25',font = ("Arial",14), height=1, width=20,fg='#A10035',background='#FEFBE7')
payment_confirm = Label(tab1, text = '',font = ("Arial",20), height=1, width=30,fg='#A10035',background='#FEFBE7')

gia_oolong.grid( row = 2, column = 2 )
gia_fruit.grid( row = 3, column = 2 )
gia_black.grid( row = 4, column = 2 )
gia_condensed.grid( row = 5, column = 2 )
payment_confirm.grid(row = 13, column = 2,pady = (10,0))


soluong_oolong = Entry(tab1,text = cal.get(), font =("Arial",14), width=10,justify ='center',fg='#A10035', textvariable = quantity_oolong)
soluong_fruit = Entry(tab1,text = "", font =("Arial",14), width=10,justify ='center',fg='#A10035',textvariable = quantity_fruit)
soluong_black = Entry(tab1,text = "", font =("Arial",14),width=10,justify ='center',fg='#A10035', textvariable = quantity_black)
soluong_condensed = Entry(tab1,text = "", font =("Arial",14),width=10,justify ='center',fg='#A10035',textvariable = quantity_condensed)



soluong_oolong.grid(row = 2, column = 3)
soluong_fruit.grid(row = 3, column = 3)
soluong_black.grid(row = 4, column = 3)
soluong_condensed.grid(row = 5, column = 3)


           
total_print = Label(tab1,text = "0", bg = 'white',relief = 'sunken', width = 30,font =("Arial",20))
total_print.grid(row = 8, column = 2, sticky = 'W',padx =(10,0))

payment_cash = Radiobutton(tab1,text = 'Cash',font =("Arial",14), width=7,justify ='center',background='#FEFBE7', value = 1,variable = method,command = click_method)
payment_card = Radiobutton(tab1,text = 'Card',font =("Arial",14), width=7,justify ='center',background='#FEFBE7',value = 2,variable = method,command = click_method)
payment_e_wallet = Radiobutton(tab1,text = 'E-Wallet',font =("Arial",14),width=7,justify ='center',background='#FEFBE7',value = 3,variable = method,command = click_method)

payment_cash.grid(row = 9, column= 2,padx = (10,0))
payment_card.grid(row = 10, column= 2,padx = (7,0))
payment_e_wallet.grid(row = 11, column= 2,padx = (40,0))

tab_control.add(tab1, text = 'Menu')

tab_control.grid()
sc.mainloop()
```
